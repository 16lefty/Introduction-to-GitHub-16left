# Externalities and Bargaining

A farmer and a rancher occupy a valley; conflict arises because their land use patterns are somewhat incompatible. The rancher makes profits $h-\frac{h^2}{2}$, where $h$ is the rancher's herd size. The farmer's profits are also influenced by herd size, and are $1 - \frac{h^2}{2}$.

1. Consider the case in which the rancher makes herd size decisions without consideration for the farmer in choosing $h$. Contrast this with the socially optimal herd size.

2.  Suppose now that the law states that the farmer has the _right to be free from crop damage by the herd_, but the the farmer and the rancher can engage in Nash bargaining over herd size and a transfer $T$ between them. First, delineate the _threat points_ and how they are determined by the law if the farmer can shut down the rancher in the event of disagreement. Then, describe the exact nature of the (Nash) bargain between the farmer and rancher.

3. Suppose that there is no law, but if an agreement is not reached, things will devolve into violence. In such a case the rancher will choose whatever herd size he or she wants to. The violence results in damages $d_r$ sustained by the rancher while the farmer sustains no damage. If the farmer and the rancher are free to bargain, what happens now?


## Answers

1. The rancher chooses herd size to maximize profits $h-\frac{h^2}{2}$. Differentiating and setting to zero gives $1-h=0$, resulting in $h^*=1$. Socially optimal herd size maximizes total returns, which are rancher and farmer returns together: $h-\frac{h^2}{2}+1-\frac{h^2}{2}$. Now, we have $1-2h=0$ as the first-order condition, so $h^{so}=\frac{1}{2}$. Clearly, the rancher chooses a herd size that is too large when acting independently.


2. First, Nash bargaining starts with Pareto-optimal outcomes (an axiom of any Nash bargain) so we know that the herd size will be chosen to maximize joint profits; to wit, when the herd size is $h=\frac{1}{2}$ (see previous), the rancher has profits $\frac{3}{8}$ and the farmer has profits $\frac{7}{8}$. If the law says the farmer has the right to be free from damage, the threat point proceeds from the rancher having profits 0 and the farmer having profits of 1, so these are the threat points. Thus, the transfer bargained over is determined by the Nash objective: $\left(\frac{3}{8}-T-0\right)\left(\frac{7}{8}+T-1\right)$. Differentiating and solving reveals $T=\frac{1}{4}$. Note that at this transfer, both the rancher and farmer gain relative to the threat points!

3. Once again, begin with the premise that the bargain will result in optimal herd size of 1/2 and profits 3/8 and 7/8 for the rancher and farmer (see previous problem). Now, however, the threat point has the rancher choosing a herd size of 1, leaving the farmer with profits of 1/2. the rancher's profits are $1/2 - d_r$. So, the Nash product to be maximized by the transfer is $\left(\frac{3}{8}-T-\left(\frac{1}{2}-d_r\right)\right)\left(\frac{7}{8}+T-\frac{1}{2}\right)$. # General Equilibrium II

Consider an exchange economy in which there are $I$ consumers, $i=1,2,3,\dots I$, each of whom has an endowment vector $\textbf{e}_i$ and consumption vector $\textbf{x}_i$. Agents' preferences are represented by a preference relation $\succ_i$, so, as usual, $\textbf{x}'_i \succ_i \textbf{x}_i$, means that agent $i$ strictly prefers consumption bundle $\textbf{x}'_i$ to $\textbf{x}_i$, and where $\textbf{x}'_i \succeq_i \textbf{x}_i$ means that $i$ considers $\textbf{x}'_i$ at least as good as $\textbf{x}_i$. You may assume that the preference relation is complete, reflexive, transitive, and that all goods are desirable.

1. Using the above notation, describe what it means for an allocation $\mathbf{x}^*$ to a) \textit{weakly} Pareto-dominate another allocation $\mathbf{x}^o$ and b) \textit{strongly} Pareto-dominate another allocation.

2. Consider the statement: ``any allocation that weakly dominates another allocation also strongly dominates another allocation.'' What assumption would allow one to assert this? Provide a proof.



## Answers

1. An allocation $\mathbf{x}'$ strongly dominates another allocation $\mathbf{x}''$ if it is better for all agents; i.e., $\mathbf{x_i}' \succ_i\mathbf{x_i}''\forall i$. An allocation $\mathbf{x}'$ weakly dominates another allocation $\mathbf{x}''$ if it is at least as good -
 $\mathbf{x_i}' \succeq_i\mathbf{x_i}''\forall i$ and for at least one $j$, we have $\mathbf{x_j}' \succeq_j\mathbf{x_j}''$.

2. The idea here is that any weakly dominant pareto allocation can be made into a strongly dominant one so long as _local nonsation_ applies. This ensures that allocations always have better allocations nearby. Why does this matter? Well, consider one allocation that weakly dominates another; in fact, suppose that at $\mathbf{x'}$ and $\mathbf{x''}$ we have that $\mathbf{x_i'}\sim_i\mathbf{x_i''}$ for all $i$ except for $j$, for whom $\mathbf{x_j'}\succ_j\mathbf{x_j''}$. Thus, agent $J$ strictly prefers $\mathbf{x'}$ while everyone else is indifferent. By local nonsation, we can "shave" off a little bit of $j$'s allocation so that we still have: $\mathbf{x_j'}-\mathbf{\epsilon}\succ_j\mathbf{x_j''}$. Also by local nonsatiation and monotonicity, we given each agent a piece of this shaving, so now we have:  $\mathbf{x_i'}+\frac{\mathbf{\epsilon}}{n-1}\succ_i\mathbf{x_i''}$ for all $i$. In the end, our allocation that weakly dominated now strongly dominates via redistribution.

# Entrepreneurs

There exists a continuum of agents who can be either entrepreneurs or workers. Workers earn a wage of $w$, while entrepreneurs hire workers. Each agent has an entrepreneurial skill level of $s$, which means that the agent can produce  
$$
    q=A \min[s, l]
$$
units of output. You might interpret this to mean that the agent can hire and use effectively $s$ workers when skill is $s$. $s$ is distributed uniformly on the unit interval, i.e., $s \sim U(0,1)$.

1. Argue that any agent who chooses to be an entrepreneur will hire exactly $s$ workers provided a small enough wage $w$, and write profits from choosing a career as an entrepreneur as a function of $s$, assuming a going equilibrium wage of $w$.

2. (Hard!) Which agents will be entrepreneurs in equilibrium, and what will be the equilibrium wage?

## Answers

1. If I choose to be an entrepreneur, I hire workers at wage $w$. Thus, profits are: $A\min[s,l]-wl$. So long as $l<s$, my profits are $l(A-w)$ so my profits are linear and increasing so long as $A>w$ (i.e., $w$ small enough). Thus, it pays to hire labor up until the point at which $l=s$, at which point my profits become $As-wl$ - now all labor does is add to my cost!

2. Previously, it was established that each entrepreneur demands $s$ units of labor, while profits from being an entrepreneur are $As-ws$ (you could add in $p$ as a price of output - doesn't matter). Who wants to be an entrepreneur? Anyone for whom $s(A-w)>w$, so $s^*=\frac{w}{A-w}$. We need supply and demand to meet as well. Supply of labor can be written as $\int_0^{s^*}ds=s^*$. These are all the people with $s$ smaller than $s^*$. Labor demand is $\int_{s^*}^1sds=\frac{1}{2}-\frac{(s^*)^2}{2}$. Substitute in and solve for $w$ and that is enough for me. You get $s^*=\sqrt{2}-1$, and the rest follows.
